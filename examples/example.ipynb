```python
import sys
import pandas as pd

sys.path.append('../')

from glmcc import GLMCC
from glmcc import spiketime_relative, glm_summary
```

```python
df = pd.read_csv('./sample_data.csv')
df.head()
```

```python
t_sp = spiketime_relative(spiketime_tar=list(df.query('neuron==1').spiketime), 
                          spiketime_ref=list(df.query('neuron==2').spiketime), window_size=50.0)

glm = GLMCC(delay=2.0)
glm.fit(t_sp)
glm.plot(t_sp, title='fitted GLM: from neuron 2 to 1')
glm_summary(glm)
```
